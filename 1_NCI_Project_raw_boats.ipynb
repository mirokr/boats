# Import libraries and dataset

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/mirokr/boats/main/boat_dataset_UTF8.csv'

data = pd.read_csv(url)
```

```python
# data.to_csv('boats_raw.csv', index=False)
```

# Remove unnamed columns

```python
#removing empty columns  
list = data.columns
list_new = []
for i in list:
  if i.startswith('Unnamed'):
    None
  else:
    list_new.append(i)
df=data[list_new]
 
```

# Sorting curency(converting to euros) and type

```python
for index in df.index:
  if type(df.loc[index,'Price']) == float:
    df.loc[index,'EditPrice'] = float('0.0')
  elif 'xa' in df.loc[index,'Price']:
    df.loc[index,'EditPrice'] = float('0.0')
  elif 'CHF' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('CHF','').replace('.','').replace(',','').replace('-',''))/1.07
  elif 'DKK' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('DKK','').replace('.','').replace(',','').replace('-',''))/7.44
  elif 'EUR' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('EUR','').replace('.','').replace(',','').replace('-',''))
  elif '£' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('£','').replace('Â','').replace('.','').replace(',','').replace('-',''))*0.85
  elif 'SEK' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('SEK','').replace('.','').replace(',','').replace('-',''))/9.99
  elif 'USD' in df.loc[index,'Price']:
    df.loc[index,'EditPrice']=float(df.loc[index,'Price'].replace('USD','').replace('$','').replace('.','').replace(',','').replace('-',''))/1.16
```

# Dropping columns that are considered not significant

```python
df.drop(['Price','Model', 'Boat name', 'Manufacturer','Hull Color','Location', 'Advertisement Date', 'Number of views last 7 days', 
           'Comments', 'Additional Comments','Category','Condition','Depth', 'Equipment','Cruising Speed','Engine Performance','Number of beds','Holding Tank','Number of Showers',
           'Number of Bathrooms','Engine', 'Max Speed'], axis=1, inplace=True)
```

# Removing zero price fields

```python
df= df[df['EditPrice'] != 0.0]
#Price rounded
for index in df.index:
  df.loc[index,'EditPrice'] = round(df.loc[index,'EditPrice'],1)
#remove outliers Price
df = df[np.abs(df.EditPrice-df.EditPrice.mean())<=(3*df.EditPrice.std())]  
```

# Sorting boat type

```python


for index in df.index:
  if type(df.loc[index,'Boat Type']) == float : df.loc[index,'Boat Type'] = float(0.0)
  elif 'Sport' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(1.0)       
  elif 'Yacht' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(2.0) 
  elif 'Trawler' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(3.0) 
  elif 'Cabin Boat' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(4.0)
  elif 'Flybridge' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(5.0)
  elif 'Fishing' in df.loc[index,'Boat Type'] : df.loc[index,'Boat Type'] = float(6.0)
  else: df.loc[index,'Boat Type'] = float(0.0)
```

# Remove TYPE of float type

```python
for index in df.index:
  if type(df.loc[index,'Type']) == float : df = df.drop(index=index)


for index in df.index:
  if 'New' in df.loc[index,'Type'] : df.loc[index,'Type'] = float(1.0)
  elif 'new' in df.loc[index,'Type'] : df.loc[index,'Type'] = float(1.0)       
  elif 'Used' in df.loc[index,'Type'] : df.loc[index,'Type'] = float(0.0)
  elif 'used' in df.loc[index,'Type'] : df.loc[index,'Type'] = float(0.0)
  elif 'Display' in df.loc[index,'Type'] : df.loc[index,'Type'] = float(1.0) 
  else: df = df.drop(index=index)
#change built year fr str to float
df['Year Built'] = df['Year Built'].astype('float')
#drop NaN values as expecting year to be important
df = df.dropna(subset=["Year Built"])
```

# Length

```python
 #all to string
for index in df.index:
  if type(df.loc[index,'Length']) == float : df.loc[index,'Length'] = str(df.loc[index,'Length'])

  #all to float without m
for index in df.index:
  if 'm' in df.loc[index,'Length']: df.loc[index,'Length'] = float(df.loc[index,'Length'].replace('m',''))  
  else: df = df.drop(index=index)
      

  #width
  #all to string
for index in df.index:
  if type(df.loc[index,'Width']) == float : df.loc[index,'Width'] = str(df.loc[index,'Width'])

  #all to float without m
for index in df.index:
  if 'm' in df.loc[index,'Width']: df.loc[index,'Width'] = float(df.loc[index,'Width'].replace('m',''))  
  else: df = df.drop(index=index)
```

# Displacement

```python
#all to string
for index in df.index:
  if type(df.loc[index,'Displacement']) == float : df.loc[index,'Displacement'] = str(df.loc[index,'Displacement'])

#all to float without m
i = "'"
for index in df.index:
  if 'kg' in df.loc[index,'Displacement']: df.loc[index,'Displacement'] = float(df.loc[index,'Displacement'].replace('kg','').replace(i,''))  
  elif '' in df.loc[index,'Displacement']: df.loc[index,'Displacement'] = float(df.loc[index,'Displacement'])
  else: df = df.drop(index=index)

#remove outliers length
df = df[np.abs(df.Length-df.Length.mean())<=(3*df.Length.std())]

#remove outliers width
df = df[np.abs(df.Width-df.Width.mean())<=(3*df.Width.std())]

#remove outliers Displacement but not removing NaN yet
df1 = pd.DataFrame()
for index in df.index:
  if pd.isna(df.loc[index,'Displacement']) or df.loc[index,'Displacement']<10:
    continue
  else:
    df1.loc[index,'Displacement_with_value']=df.loc[index,'Displacement']

meanD = df1.Displacement_with_value.mean()
stdD = df1.Displacement_with_value.std()

#removing outliers on available displacement values
for index in df.index:
  if pd.isna(df.loc[index,'Displacement'])or df.loc[index,'Displacement']<10:
    continue
  else:
    if np.abs(df.loc[index,'Displacement']-meanD)>(3*stdD):
      df = df.drop(index=index)
    

#a lot of NaN in displacement
#lets see the ratio between length/width and displacement considering length(ships weight per length)
ratio = 0
count = 0
ratio1 = 0
count1 = 0
ratio2 = 0
count2 = 0
for index in df.index:
  if pd.isna(df.loc[index,'Displacement']) or df.loc[index,'Displacement']<10:
    continue
  else: 
    if df.loc[index,'Length']<=7:
      ratio += df.loc[index,'Displacement']/(df.loc[index,'Length']+df.loc[index,'Width'])
      count+=1
    elif df.loc[index,'Length']>7 and df.loc[index,'Length']<=15:
      ratio1 += df.loc[index,'Displacement']/(df.loc[index,'Length']+df.loc[index,'Width'])
      count1+=1
    elif df.loc[index,'Length']>15:
      ratio2 += df.loc[index,'Displacement']/(df.loc[index,'Length']+df.loc[index,'Width'])
      count2+=1
    


avg_ratio = ratio / count
avg_ratio1 = ratio1 / count1
avg_ratio2 = ratio2 / count2

#applying ratio on NaN displacement to replace NaN

for index in df.index:
  if pd.isna(df.loc[index,'Displacement']) or df.loc[index,'Displacement']<10:
    if df.loc[index,'Length']<=7:
      df.loc[index,'Displacement'] = round((avg_ratio * (df.loc[index,'Length']+df.loc[index,'Width'])),1)
    elif df.loc[index,'Length']>7 and df.loc[index,'Length']<=15:
      df.loc[index,'Displacement'] = round((avg_ratio1 * (df.loc[index,'Length']+df.loc[index,'Width'])),1)
    elif df.loc[index,'Length']>15:
      df.loc[index,'Displacement'] = round((avg_ratio2 * (df.loc[index,'Length']+df.loc[index,'Width'])),1)
```

```python
# df = pd.read_csv('/content/boats_price_up_to_People.csv')
#People - after examining best to just replace NaN with 0
for index in df.index:
  if pd.isna(df.loc[index,'People']) : df.loc[index,'People']=0.0
  else: df.loc[index,'People']=float(df.loc[index,'People'])

#Cabins - after examining best to just replace NaN and negative values with 0
for index in df.index:
  if pd.isna(df.loc[index,'Cabins']) or float(df.loc[index,'Cabins'])<0 : df.loc[index,'Cabins']=0.0
  else: df.loc[index,'Cabins']=float(df.loc[index,'Cabins'])

#Toilets - after examining best to just replace NaN and negative values with 0
for index in df.index:
  if pd.isna(df.loc[index,'Toilets']) or float(df.loc[index,'Toilets'])<0 : df.loc[index,'Toilets']=0.0
  else: df.loc[index,'Toilets']=float(df.loc[index,'Toilets'])
    

#Material 
for index in df.index:
  if type(df.loc[index,'Material']) == float : df.loc[index,'Material'] = float(0.0)
  elif df.loc[index,'Material'].__len__()>15 : df = df.drop(index=index) #no material with name longer then 15 char
  elif 'Thermoplastic' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(1.0)
  elif 'Aluminium' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(2.0)
  elif 'GRP' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(3.0)
  elif 'PVC' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(4.0)
  elif 'Plastic' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(5.0)
  elif 'Wood' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(6.0)
  elif 'Steel' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(7.0)
  elif 'Hypalon' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(8.0)
  elif 'Carbon Fiber' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(9.0)
  elif 'Rubber' in df.loc[index,'Material'] : df.loc[index,'Material'] = float(10.0)
  else: df.drop(index=index)


#water
for index in df.index:
  if pd.isna(df.loc[index,'Water']) : df.loc[index,'Water'] = float(0)
  elif 'l' in df.loc[index,'Water'] : df.loc[index,'Water'] = float(df.loc[index,'Water'].replace('l', ''))
```
